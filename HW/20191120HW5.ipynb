```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)

f = lambda x,y:(7 - x / 2 + x**4 + y**10)*np.exp(-0.1*x**2 - 0.1*y**2)

plt.contourf(X, Y,f(X, Y) , 5, alpha = 1, cmap = 'rainbow')
C = plt.contour(X, Y, f(X, Y), 5, colors = 'black', linewidth = 0.3)
plt.clabel(C, inline = True, fontsize =12)
```
