```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk
import time 
import datetime as dt
```

```python
window = tk.Tk()
window.title("Date count down")
window.geometry('400x300')
window.configure(background='white')


def datecountdown():
    today = dt.date.today()
    thedate = date_entry.get()
    year = int(year_entry.get())
    month = int(month_entry.get())
    day = int(day_entry.get())
    date = dt.date(year,month,day)
    D = date - today
    result =  '距離'+thedate+'還剩下'+str(D.days)+'天'
    result_label.configure(text=result)

header_label = tk.Label(window, text='倒數計時器')
header_label.pack()

date_frame = tk.Frame(window)
date_frame.pack(side=tk.TOP)
date_label = tk.Label(date_frame, text='重要日期')
date_label.pack(side=tk.LEFT)
date_entry = tk.Entry(date_frame)
date_entry.pack(side=tk.LEFT)

year_frame = tk.Frame(window)
year_frame.pack(side=tk.TOP)
year_label = tk.Label(year_frame, text='年')
year_label.pack(side=tk.LEFT)
year_entry = tk.Entry(year_frame)
year_entry.pack(side=tk.LEFT)

month_frame = tk.Frame(window)
month_frame.pack(side=tk.TOP)
month_label = tk.Label(month_frame, text='月')
month_label.pack(side=tk.LEFT)
month_entry = tk.Entry(month_frame)
month_entry.pack(side=tk.LEFT)

day_frame = tk.Frame(window)
day_frame.pack(side=tk.TOP)
day_label = tk.Label(day_frame, text='日')
day_label.pack(side=tk.LEFT)
day_entry = tk.Entry(day_frame)
day_entry.pack(side=tk.LEFT)

result_label = tk.Label(window)
result_label.pack()

countdown_btn = tk.Button(window, text='倒數計時',command=datecountdown)
countdown_btn.pack()





window.mainloop()
```
