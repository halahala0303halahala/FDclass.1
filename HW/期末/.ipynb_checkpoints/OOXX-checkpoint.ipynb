```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact_manual
```

```python
def draw_X(i):
    a,b = site[i]
    plt.plot((a-2,a+2),(b+2,b-2),'b')
    plt.plot((a-2,a+2),(b-2,b+2),'b')

def draw_O(i):
    a,b = site[i]
    x = np.arange(a-2, a+2, 0.01)
    y1 = b + np.sqrt(4 - (x - a)**2)
    y2 = b - np.sqrt(4 - (x - a)**2)
    plt.plot(x, y1,'r')
    plt.plot(x, y2,'r')

def OOXX():
    plt.figure(figsize=(5,5))
    plt.plot((2,2),(0,12),'k')
    plt.plot((-2,-2),(0,12),'k')
    plt.plot((6,-6),(8,8),'k')
    plt.plot((6,-6),(4,4),'k')
    draw_X(input(''))
    draw_O(input(''))
    draw_X(input(''))
    draw_O(input(''))
    draw_X(input(''))
    draw_O(input(''))
```
