```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import geopandas as gpd
```

```python
town = gpd.read_file('TOWN_MOI_1081121.shp',encoding='utf-8')
```

```python
import matplotlib.pylab as plb
```

```python
plb.rcParams['figure.figsize'] = (170, 170)
```

```python
df= pd.read_csv("AQI_20191202161010.csv")
df.head()
```

```python
df1 = df[['County', 'AQI']]
df1.head(10)
```

```python
df_aqi = df1.groupby(['County']).mean().round(2)
df_aqi.head(10)
```

```python
df_town = pd.merge(town,df_aqi,left_on=('COUNTYNAME'),right_on=('County'))
df_town.head()
```

```python
df_town.plot(column='AQI', k=50, cmap='cool');
```
